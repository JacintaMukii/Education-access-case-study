**Education Facilities Dataset:**

4. **Level_**: School level (e.g., primary).

7. **Type1, Type2, Type3**: Type of school (e.g., day, boarding).
8. **PupilTeach**: Student-to-teacher ratio.
9. **ClassrmRat**: Classroom-to-student ratio.
10. **ToiletRati**: Toilet-to-student ratio.
11. **No_Classrm**: Number of classrooms.
12. **BoysToilet, GirlsToilet**: Number of toilets for boys and girls.
13. **TeachersTo**: Teacher-to-student ratio.
14. **TotalToile**: Total number of toilets in the school.
15. **TotalBoys, TotalGirls**: Total number of boys and girls in the school.
16. **TotalEnrol**: Total school enrollment.
17. **GO_KTSC_M, GOK_TSC_F**: Government-paid male and female teachers.
18. **authorityM, authorityF**: Male and female school authorities.
19. **PTA_BOG_M, PTA_BOG_F**: Male and female members of the Parent-Teacher Association.
20. **OthersM, OthersF**: Other staff members (male and female).
21. **NoTeaching, NonTeachin**: Number of non-teaching and teaching staff.
22. **Province, District, Division**: Geographic divisions in Kenya.
23. **Location, Costituenc**: School's specific location and constituency.
24. **Latitude, Longitude**: Geographic coordinates.

**Kenya Population and Housing Census data:**
This dataset provides Distribution of Population Age 3 Years and Above by School Attendance Status, Area of Residence, Sex, County and Sub-County 2019 Census.

The dataset has both numerical and categorical values and has 1183 rows and 13 columns.

1. **County/Sub-County**: Geographic regions being analyzed.
2. **Total**: Total number of individuals in the dataset.
3. **At School/Learning Institution**: Students currently attending school.
4. **Left School After Completion**: Students who finished their education and left.
5. **Left School Before Completion**: Students who dropped out before finishing.
6. **Never Been to School**: Individuals who have never attended school.
7. **Don't Know**: Unknown educational status.
8. **Not Stated**: Missing or unreported educational status.

### 4.1 Dropping irrelevant columns
After loading the data, drop the columns that are not necessary for the analysis.

For the Education Facilities Dataset:

Kenya Primary Schools Dataset

```python
# Drop irrelevant columns
df_cleaned = df.drop(columns=[
    'X', 'Y', 'Type1','Type2','Type3', 'Province', 'Division', 'OthersF', 'OthersM','PTA_BOG_F', 'PTA_BOG_M', 'GO_KTSC_M', 'GOK_TSC_F' , 'authorityF', 'authorityM'
])
df_cleaned.head()
```

For the Kenya Population and Housing Census dataset

```python
df1.columns
```

```python
# Dropping irrelevant columns
df1_cleaned = df1.drop(columns=[
    'Don\'t Know', 'Not Stated', 'Unnamed: 11'
])
df1_cleaned.head()
```

### 4.2 Renaming the columns for both datasets

For the Education Facilities Dataset:

```python
# Renaming the 'District' column to 'County/Sub-County'
df_cleaned = df_cleaned.rename(columns={'District': 'County/ Sub-County'})
df_cleaned.head()
```

Kenya Population and Housing Census dataset:

```python
df1_cleaned.columns
```

```python
#Renaming the unnamed columns
rename_mapping = {
    'Unnamed: 3': 'At School/ Learning Institution %',
    'Unnamed: 5': 'Left School/ Learning Institution After Completion %',
    'Unnamed: 7': 'Left School/ Learning Institution Before Completion %',
    'Unnamed: 9': 'Never Been to School/ Learning Institution %',

}
df1_cleaned.rename(columns=rename_mapping, inplace=True)
df1_cleaned.columns
```

### 4.3 Change Data Types
For Education Dataset:

Convert the numerical columns to appropriate numeric types, and categorical columns to "category":

```python
# Converting Constituency to categorical in education dataset
df_cleaned['Constituency'] = df_cleaned['Costituenc'].astype('category')

# Converting TotalEnrol (total enrollment) to numeric in education dataset
df_cleaned['TotalEnrol'] = pd.to_numeric(df_cleaned['TotalEnrol'], errors='coerce')

```

```python
df_cleaned.info()
```

For the Population Dataset:

Convert the numerical columns to appropriate numeric types, and categorical columns to "category":

```python
# Converting numerical columns in population dataset to numeric
numerical_cols = [
    'At School/ Learning Institution',
    'At School/ Learning Institution %',
    'Left School/ Learning Institution After Completion',
    'Left School/ Learning Institution After Completion %',
    'Left School/ Learning Institution Before Completion',
    'Left School/ Learning Institution Before Completion %',
    'Never Been to School/ Learning Institution',
    'Never Been to School/ Learning Institution %',

]

for col in numerical_cols:
    df1_cleaned[col] = pd.to_numeric(df1_cleaned[col], errors='coerce')  # Invalid entries become NaN

# Converting County/Sub-County to categorical
df1_cleaned['County/ Sub-County'] = df1_cleaned['County/ Sub-County'].astype('category')
```

### 4.4 Checking for missing values

For the Education dataset:

```python
df_cleaned.isnull().sum()
```

```python
#Handling the missing values
df_cleaned.dropna(inplace=True)
df1_cleaned.isnull().sum()
```

For the Population dataset:

```python
df1_cleaned.isnull().sum()
```

```python
# Filling missing values in numerical columns with the mean in the Population dataset
for col in numerical_cols:
    df1_cleaned[col].fillna(df1_cleaned[col].mean(), inplace=True)

# Filling missing values in 'County/ Sub-County' with the mode
df1_cleaned['County/ Sub-County'].fillna(df1_cleaned['County/ Sub-County'].mode()[0], inplace=True)
df1_cleaned.isnull().sum()
```

Both datasets have gone through the preprocessing stage.

### 4.5 Merging the two datasets


```python
df_cleaned.columns
```

```python
merged_data = pd.merge(
    df_cleaned,
    df1_cleaned,
    how='outer',
    on='County/ Sub-County'  # Common key column
)
merged_data.copy()
# Displaying merged dataset
print(merged_data.shape)
print(merged_data.head())
```

```python
merged_data.columns
```

```python
#Detecting and handling missing values
print(merged_data.isnull().sum())
```

```python
#Handling the missing values in the dataset
for index, row in merged_data.iterrows():
    # Iterating through each column in the current row
    for col in merged_data.columns:
        # Checking if the current cell is null
        if pd.isnull(row[col]):
            if col in ['TotalEnrol', 'PupilTeach', 'ClassrmRat', 'ToiletRati','TotalBoys','TotalGirls']:
                #For numerical columns, fill with the mean of the column
                merged_data.loc[index, col] = merged_data[col].mean()
            elif col == 'Constituency':
                # For categorical columns, fill with the mode of the column
                merged_data.loc[index, col] = merged_data[col].mode()[0]
            elif col in numerical_cols:  # Numerical columns from the population dataset
                merged_data.loc[index,col] = merged_data[col].mean()
            elif col == 'County/ Sub-County': #Categorical column
                merged_data.loc[index, col] = merged_data[col].mode()[0]

            else:
                merged_data.loc[index, col] = 0
# Verifying if there are any remaining null values after processing
print(merged_data.isnull().sum())
```

The two datasets have been preprocessed and merged, ready for visualizations.

### 6.1 Enrollment Rate by Region

```python
#Plotting the enrollment rate in for the top 5 columns
if 'Enrollment Rate' not in merged_data.columns:
    merged_data['Enrollment Rate'] = (merged_data['At School/ Learning Institution'] / merged_data['TotalEnrol']) * 100
# Sorting the data by Enrollment Rate
sorted_df = merged_data.sort_values(by='Enrollment Rate', ascending=False)
# Plotting the top regions using 'At School/ Learning Institution %' as enrollment rate
plt.figure(figsize=(8, 6))
ax = sns.barplot(data=sorted_df.head(6), x='County/ Sub-County', y='At School/ Learning Institution %')
plt.xticks(rotation=45)
plt.title('Top 5 Counties/Sub-Counties by Enrollment Rate (% of Population)')

# Adding the percentages on the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

The bar plot shows that Nairobi is among the top five counties with the highest enrollment rates, along with Lugari, Vihiga, Machakos, and Kitui. This highlights Nairobi's significant progress in increasing school enrollment, though further analysis is needed to ensure this growth is matched by improvements in infrastructure and resource allocation.
